```python
X = df.drop(['1','0','g'],axis =1 )
```

```python
X.head()
```

```python
y = df['g']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
y_test = y_test.replace(['g','b'],[0,1])
y_train = y_train.replace(['g','b'],[0,1])
```

```python
model = Sequential() 
model.add(Dense(8,input_shape = (32,),activation="relu"))
model.add(Dense(units = 10,kernel_regularizer=regularizers.l2(0.01),activation= "relu"))
model.add(Dense(units = 16,activation= "relu"))
model.add(Dense(units = 1,activation= "sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer="sgd",metrics=['accuracy'],loss = 'binary_crossentropy')
```

```python
model.fit(X_train,y_train,batch_size=8,epochs=20,verbose=1,validation_data=(X_test,y_test))
```

```python
model.evaluate(X_train,y_train)
```

```python
model.evaluate(X_test,y_test)
```
